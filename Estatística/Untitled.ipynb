### Parte 1
Aqui iremos trabalhar os dados

```python
df.Date = pd.to_datetime(df.Date) # convert to date format

type(df["White CP Loss List"][0][0]) # verify the variable type in the list
```

As seen above, it is necessary to convert data of type str to numerical values.

```python
def transform_list_str_to_int(df_column): # função que transforma lista de strings em lista de ints
    return df_column.apply(lambda x: [eval(i) for i in x[1:-1].replace(" ", "").split(',')])

df['Evaluations List'] = transform_list_str_to_int(df['Evaluations List'])
df['White CP Loss List'] = transform_list_str_to_int(df['White CP Loss List'])
df['Black CP Loss List'] = transform_list_str_to_int(df['Black CP Loss List'])

type(df['White CP Loss List'][0][0])
```

```python
z = df.Result.apply(lambda x: x.replace("1/2", "0.5").replace("1", "2").replace("0.5", "1")) # transformando em inteiros
k = pd.DataFrame(z.str.split('-', expand=True).values, columns=['Result White', 'Result Black']) # separando em duas colunas
df['Result White'], df['Result Black'] = k['Result White'].astype(int), k['Result Black'].astype(int) # add ao df e convertendo p int
df.drop('Result', axis=1, inplace=True) # removendo  coluna result
df.head(3)
```

__Standard Deviation CP Loss__

```python
import numpy as np
def get_the_means(column): return column.apply(lambda x: np.mean(x)).round(2) # get the average of a list
def get_the_stds(column): return column.apply(lambda x: np.std(x)).round(2) # get the std of a list

df["White Av CP Loss"] = get_the_means(df['White CP Loss List']) # substitui o valor pelo mais aproximado
df["Black Av CP Loss"] = get_the_means(df['Black CP Loss List'])

df["White Std CP Loss"] = get_the_stds(df['White CP Loss List'])
df["Black Std CP Loss"] = get_the_stds(df['Black CP Loss List']); df.head(3)
```

## Get the start date for each player

```python
player_names = pd.array(df_date.Name.value_counts().index)#; player_names
```

```python
first_game_data = []
for i in range(len(player_names)):
    first_game_data.append(df_date[df_date.Name == player_names[i]].Date.values[0]) # date from the first game
first_game_data = pd.array(first_game_data)#; first_game_data
```

```python
df_1st_play = pd.DataFrame({'Name':player_names, 'First Game Date':first_game_data}); df_1st_play
```

```python
df_1st_play[df_1st_play['Name'] == 'Rausis, Igors']['First Game Date'].values[0]
```

```python
#df_date.Name
def get_first_date(player_name): # get the day of the first match
    return df_1st_play[df_1st_play['Name'] == player_name]['First Game Date'].values[0]
```
